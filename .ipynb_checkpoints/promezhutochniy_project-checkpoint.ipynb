# Task 2: Сколько заказов в месяц в среднем не доставляется по разным причинам (вывести детализацию по причинам)?

## Выведем недоставленные заказы

```python
not_delivered_orders = orders.query("order_status != 'delivered'")
```

```python
not_delivered_orders.head()
```

```python
# Добавим новый столбец с месяцом в котором был сделан заказ
not_delivered_orders['order_month'] = not_delivered_orders.order_purchase_timestamp.dt.month
```

```python
# Добавим новый столбец с годом в котором был сделан заказ
not_delivered_orders['order_year'] = not_delivered_orders.order_purchase_timestamp.dt.year
```

Недоставленными заказами будем считать со статусами canceled и unavailable, так как остальные могут находиться в процессе формирования или в работе

```python
# Преобразуем наш дата фрейм к этим заказам
not_delivered_orders = not_delivered_orders.query("order_status in ('canceled', 'unavailable')")
```

```python
# Сгруппируем данные по заказам со статусом canceled по году и месяцу, в который был сделан заказ и высчитаем среднее
canceled = not_delivered_orders.query("order_status == 'canceled'")\
                               .groupby(['order_year', 'order_month'], as_index=False)\
                               .agg({'order_id':'count'})
```

```python
canceled_per_month = canceled.order_id.mean().round(2)
```

```python
canceled_per_month
```

```python
# Проделаем то же самое с заказами со статусом unavailable
unavailable = not_delivered_orders.query("order_status == 'unavailable'")\
                               .groupby(['order_year', 'order_month'], as_index=False)\
                               .agg({'order_id':'count'})
```

```python
unavailable_per_month = unavailable.order_id.mean().round(2)
```

```python
unavailable_per_month
```

# Task 4: Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)? Не стоит забывать, что внутри месяца может быть не целое количество недель. Например, в ноябре 2021 года 4,28 недели. И внутри метрики это нужно учесть.

```python
# Сформируем датафрейм
purchase_per_week = customers.merge(delivered, on='customer_id')
```

```python
# Для точности данных выберем только неотмененные заказы
purchase_per_week = purchase_per_week.query("order_status != 'canceled'")
```

```python
# Введем новую колонку с периодом покупки
purchase_per_week['month_of_the_year'] = purchase_per_week.order_purchase_timestamp.dt.to_period("M")
```

```python
# Создадим новый df с группировкой по пользователю и периоду покупки
purchase_per_week = purchase_per_week.groupby(['customer_unique_id', 'month_of_the_year'], as_index=False)\
                                     .agg({'order_id':'count'})
```

```python
purchase_per_week
```

```python
# Введем новую колонку с количеством недель в месяце
purchase_per_week['weeks_in_month'] = purchase_per_week.month_of_the_year.dt.days_in_month / 7
```

```python
# Введем новую колонку с количесвом заказов в неделю
purchase_per_week['purchases_in_week'] = purchase_per_week.order_id / purchase_per_week.weeks_in_month
```

```python
purchase_per_week
```

```python
# Вычислим среднее по каждому пользователю
purchase_per_week = purchase_per_week.groupby('customer_unique_id', as_index=False).agg({'purchases_in_week':'mean'})
```

# Task 5: Используя pandas, проведи когортный анализ пользователей. В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.

```python
# Сформируем общий дата фрейм
df_prom = customers.merge(delivered, on= 'customer_id')
```

```python
df = df_prom.merge(items, on='order_id')
```

```python
df.head()
```

```python
# Сформируем периоды покупок
df['period'] = df.order_approved_at.dt.strftime('%Y-%m')
```

```python
df.head()
```

```python
# Сформируем период первой покупки
df.set_index('customer_unique_id', inplace=True)
```

```python
# Отсортируем данные с января по декабрь
df = df.query("period >= '2017-01' and period < '2018-01'")
```

```python
df['first_purchase_period'] = df.groupby(level=0)['order_approved_at'].min().dt.strftime('%Y-%m')
```

```python
# Создадим когорты
df.reset_index(inplace=True)
```

```python
cohorts = df.groupby(['first_purchase_period', 'period']).agg({'customer_unique_id':'nunique', 'order_id':'count', 'price':'sum'})
```

```python
cohorts.head()
```

```python
# Переименуем колонки
cohorts.rename(columns={'customer_unique_id':'total_customers', 'order_id':'total_orders', 'price':'total_price'}, inplace=True)
```

```python
cohorts
```

```python
# Введем формулу для упорядочивания когортов
def cohort_period(df_exp):
    df_exp['cohort_period'] = np.arange(len(df_exp)) + 1
    return df_exp
```

```python
# Применим данную формулу для наших когортов
cohorts = cohorts.groupby(level=0).apply(cohort_period)
```

```python
cohorts
```

```python
# Переиндексируем наш df по периоду когорта и периоду первой покупки
cohorts.reset_index(inplace=True)
```

```python
cohorts.set_index(['cohort_period', 'first_purchase_period'], inplace=True)
```

```python
cohorts
```

```python
# Выделим значения для сравнения для каждого периода первой покупки
cohort_period_size = cohorts.groupby(level=1)['total_customers'].first()
```

```python
cohort_period_size
```

```python
# Преобразуем нашу таблицу, чтобы периоду когортов являлись столбцами, индексами являлись периоды первой покупки, а значениями
# являлось количество клиентов
```

```python
cohorts['total_customers'].unstack(0)
```

```python
# Вычисляем customers retention
```

```python
customers_retention = cohorts['total_customers'].unstack(0).divide(cohort_period_size, axis=0)
```

```python
customers_retention
```

# Ответ: максимальный retention на 3-й месяц наблюдается в когорте пользователей 2017-06

```python
customers_retention.loc[customers_retention[4].idxmax()]
```

# Task 6: Выполнить RFM сегментацию пользователей. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время.

```python
# Сформируем общий дата фрейм
df_prom = customers.merge(delivered, on= 'customer_id')
```

```python
df = df_prom.merge(items, on='order_id')
```

```python
# Определим текущую дату
current_date = df.order_approved_at.max() + pd.to_timedelta(1, 'D')
```

```python
# Создадим новый df с характеристиками в задании по уникальному идентификатору пользователя
rfm_df = df.groupby('customer_unique_id')\
           .agg({'order_approved_at':lambda x: (current_date - x.max()).days, 'order_id':'count', 'price':'sum'})
```

```python
# Переименуем колонки
rfm_df.rename(columns={'order_approved_at':'recency', 'order_id':'frequency', 'price':'momentary_revenue'}, inplace=True)
```

```python
rfm_df
```

```python
# Введем диапозоны распределения искомых величин в виде 4 квантилей
quantiles = rfm_df.quantile([0.2, 0.4, 0.6, 0.8])
```

```python
quantiles = quantiles.to_dict()
```

```python
# Применим функцию для определения рейтинга
def score(x,p,d):
     if x <= d[p][0.20]:
         return str(1)
     elif x <= d[p][0.40]:
         return str(2)
     elif x <= d[p][0.60]: 
         return str(3)
     elif x<=d[p][0.80]:
         return str(4)
     else:
         return str(5)
```

```python
# Создадим новые столбцы с рейтингом величины конкретного пользователя с помощью функции, написанной раннее
rfm_df['r_quantile'] = rfm_df.recency.apply(score, args=('recency', quantiles))
```

```python
rfm_df['f_quantile'] = rfm_df.frequency.apply(score, args=('frequency', quantiles))
```

```python
rfm_df['m_quantile'] = rfm_df.momentary_revenue.apply(score, args=('momentary_revenue', quantiles))
```

```python
rfm_df['score'] = rfm_df.r_quantile + rfm_df.f_quantile + rfm_df.m_quantile
```

```python
# Введем сегменты
segt_map = {
    r'[1-2][1-2]': 'ненадежный',
    r'[1-2][3-4]': 'в зоне риска потери',
    r'[1-2]5': 'не потеряем (возможно)',
    r'3[1-2]': 'находится с в стадии сна',
    r'33': 'требует внимания',
    r'[3-4][4-5]': 'постоянный клиент',
    r'41': 'обещанный покупатель',
    r'51': 'новый клиент',
    r'[4-5][2-3]': 'потенциально постоянный клиент',
    r'5[4-5]': 'самый постоянный клиент'
}
```

```python
rfm_df['segment'] = rfm_df['r_quantile'] + rfm_df['f_quantile']
```

```python
rfm_df['segment'] = rfm_df['segment'].replace(segt_map, regex=True)
```
